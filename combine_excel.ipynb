```python
import pandas as pd

excel_file = 'managers.xlsx'
to_out = 'Managers.xlsx'

df = pd.read_excel(excel_file)
```

```python
xlsx = pd.ExcelFile(excel_file)
excel_sheets = []
for sheet in xlsx.sheet_names:
    excel_sheets.append(xlsx.parse(sheet))
# df = pd.concat(excel_sheets, ignore_index=True)
df = pd.concat(excel_sheets, ignore_index=False)
```
